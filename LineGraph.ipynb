```python
import pandas as pd

y ={'f(x)':[2,4,10,8],'g(x)':[3,-2,15,12]}
x=[1,2,3,4]

graph =pd.DataFrame(y,x)
graph.plot(grid=True, title='Graph', xlabel='X axis', ylabel= 'Y axis',xlim = (1,4))
```
